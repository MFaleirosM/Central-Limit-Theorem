
<!-- # Parâmetros e estatísticas -->

Os parâmetros se referem à população enquanto que as estatísticas se referem às amostras. <br><br>





**Alguns parâmetros populacionais**<br><br>
Média populacional:
$$
\mu=\frac{\sum_{i=1}^{N}X_i}{N}
$$
Variância populacional:
$$
\sigma^2=\frac{\sum_{i=1}^{N}(X_i-\mu)^2}{N}
$$

```python
# Cálculo da média e variância populacionais (já feito em duas células acima!)

media_populacional = dados['Populacao'].mean()
variancia_populacional = dados['Populacao'].var(ddof=0)

print('A média populacional é: ', round(media_populacional, 4), '\n')
print('A variância populacional é:', round(variancia_populacional, 5))
```

**Alguns estimadores amostrais** <br>
Média amostral:
$$
\hat\mu=\frac{\sum_{i=1}^{N}X_i}{n}
$$
Obs.: outra nomenclatura muito comum para $\hat\mu$ é $\bar{X}$. <br><br>
Variância amostral:
$$
\hat\sigma^2=\frac{\sum_{i=1}^{n}(X_i-\mu)^2}{n-1}
$$


```python
# Extraindo uma amostra aleatória com reposição

amostra = dados['Populacao'].sample(10, replace=True)
print(amostra.to_string(),'\n')
```

```python
# Cálculo da média amostral e desvio padrão amostral para a amostra extraída acima

media_amostral = amostra.mean()
variancia_amostral = amostra.var(ddof=1)

print('A média amostral é', round(media_amostral,2))
print('A variância amostral é', round(variancia_amostral,6))
```

**Vício**<br>
Um estimador é não viciado ou não viesado para um parâmetro $\theta$ se $E(\hat\theta)=\theta$. Em outras palavras, um estimador é não viciado se o seu valor esperado coincide com o parâmetro de interesse. Podemos pensar no valor esperado $E(\hat\theta)$ como sendo aproximadamente a média de $\hat\theta$ calculada para uma grande quantidade de amostras. Perceba também que esta igualdade não depende do tamanho $n$ da amostra, ou seja, deve ocorrer para todo $n$ fixado.

**Consistência**<br>
Um estimador $\hat\theta$ é consistente, se, à medida que o tamanho $n$ da amostra aumenta, seu valor esperado converge para o parâmetro de interesse e sua variância converge para zero. Ou seja, $\hat\theta$ é consistente se as duas propriedades seguintes são satisfeitas:<br><br>
$$
(i)\quad \lim_{n\rightarrow\infty}E(\hat\theta)=\theta
$$
<br> e <br>
$$
(ii)\quad \lim_{n\rightarrow\infty}Var(\hat\theta)=0.
$$


## Valor esperado do estimador $\hat\sigma$ (variância), em outras palavras: $E(\hat\sigma)$.

```python
# Coletando as amostras de tamanho n=10 e preenchendo uma lista de valores de todas as amostras chamada de base_amostras
base_amostras = []
for i in range(0, qtd_amostras):
  base_amostras.append(dados['Populacao'].sample(10, replace=True))
# len(base_amostras) #numero de amostras
# len(base_amostras[0]) #numero de valores em cada amostra
```

```python
# Explicitando as amostras coletadas com cada amostra em uma linha
base_amostras = np.array(base_amostras)
base_amostras;
```

```python
# Explicitando as amostras coletadas com cada amostra em uma coluna
base_amostras = np.array(base_amostras).T
base_amostras;
# np.array(base_amostras).T.tolist()
```

```python
# Exibindo as amostras coletadas em uma tabela onde cada coluna contém os dados de uma amostra
base_amostras = pd.DataFrame(np.array(base_amostras))
print(base_amostras)
```

```python
# Obtendo as variâncias amostrais para cada amostra
colecao_de_variancias_amostrais = [base_amostras[i].var(ddof=1) for i in range(qtd_amostras)]
colecao_de_variancias_amostrais = pd.DataFrame(colecao_de_variancias_amostrais)
colecao_de_variancias_amostrais.T
```

```python
# Comparando a média das variâncias amostrais coletadas com a variância populacional
media_variancias_amostrais = colecao_de_variancias_amostrais.mean()
print(' Média das variâncias da coleção de amostras:', round(media_variancias_amostrais.squeeze(), 6),'\n\n',\
      'Variância populacional:', round(variancia_populacional,6))
```

### Gráfico de densidade da distribuição das variâncias amostrais comparado com a curva qui-quadrado

```python
# Montando um gráfico do tipo histograma para visualizar a curva de densidade das variâncias amostrais
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import scipy.stats as stats
import math

# Número de graus de liberdade
k = n-1
tabela_variancias_amostrais_modificada = pd.DataFrame()
tabela_variancias_amostrais_modificada['Variâncias amostrais modificadas'] = tabela_variancias_amostrais['Variâncias amostrais']*k/2**2

h = tabela_variancias_amostrais_modificada['Variâncias amostrais modificadas'].hist(bins=100, grid=True, figsize=(12,8), color='#86bf91', rwidth=0.9, edgecolor="black", density=1)
# Ver "Colour picker" no Google para escolher a cor de preferência
# Não confundir densidade com frequência relativa

# Inserindo legendas no histograma
h.set_xlabel("Variâncias amostrais", labelpad=20, weight='bold', size=14)
h.set_ylabel("Frequência absoluta", labelpad=20, weight='bold', size=14)

# Adicionando título ao gráfico
h.set_title(r"Histograma de variâncias amostrais modificada, ou seja, $\frac{S^2\cdot (n-1)}{\sigma^2}$", weight='bold', size=20, color = 'blue')

# Controlando o tamanho da fonte dos eixos
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Adicionando a curva de densidade ajustada ao histograma
tabela_variancias_amostrais_modificada['Variâncias amostrais modificadas'].plot.kde(color='red')

# Média e desvio padrão para a curva qui-quadrado a ser plotada
mu, sigma = k, math.sqrt(2*k)

# Construindo a curva normal da população
x = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)
plt.plot(x, stats.chi2.pdf(x, k), alpha=0.8, linewidth=2.0)

# Adicionando linhas de grid
plt.grid(axis='y', alpha=0.3, color ='black')
plt.grid(axis='x', alpha=0.3, color ='black')

# Restringindo os limites das abscissas do gráfico
plt.xlim([-2, 40])
plt.show()
```
